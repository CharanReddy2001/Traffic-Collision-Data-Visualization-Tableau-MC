## 1 Introduction
The Montgomery County Traffic Collision Dataset is a comprehensive compilation of 172,105 records spanning 43 columns. It offers a detailed overview of traffic accidents occurring on county and local roads within Montgomery County. Sourced from the Maryland State Police's Automated Crash Reporting System, this dataset provides insights into collision occurrences and involved drivers. With easy access to information on incident details and preliminary classifications, analysts and policymakers can utilize this extensive dataset to identify patterns and develop strategies aimed at enhancing road safety measures throughout the area.

Analyzing the Montgomery County Traffic Collision Dataset can help save lives by identifying accident hotspots, enhancing road safety measures, and influencing policy choices. It encourages community collaboration, improves resource allocation, and promotes research into new traffic collision prevention solutions.

**Attribute Details:**

- Report Number: Unique identifier for each collision report.
- Local Case Number: Local case number assigned to the collision by the reporting agency.
- Agency Name: Name of the law enforcement agency reporting the collision.
- ACRS Report Type: Type of crash report according to the Automated Crash Reporting System.
- Crash Date/Time: Date and time when the collision occurred.
- Route Type: Type of route where the collision occurred (e.g., Maryland State, County).
- Road Name: Name of the road where the collision occurred.
- Cross-Street Type: Type of street intersecting with the road where the collision occurred.
- Cross-Street Name: Name of the intersecting street.
- Off-Road Description: Description of the location off the road where the collision occurred.
- Municipality: Municipality where the collision occurred.
- Related Non-Motorist: Involvement of non-motorists (e.g., pedestrians, cyclists) in the collision.
- Collision Type: Type of collision (e.g., rear-end, head-on).
- Weather: Weather conditions at the time of the collision.
- Surface Condition: Condition of the road surface at the time of the collision.
- Light: Lighting conditions at the time of the collision.
- Traffic Control: Type of traffic control at the collision site (e.g., traffic signal).
- Driver Substance Abuse: Whether the driver was under the influence of substances.
- Non-Motorist Substance Abuse: Whether the non-motorist involved was under the influence of substances.
- Person ID: Unique identifier for each person involved in the collision.
- Driver At Fault: Whether the driver was at fault for the collision.
- Injury Severity: Severity of injuries sustained in the collision.
- Circumstance: Circumstances surrounding the collision.
- Driver Distracted By: Factors distracting the driver at the time of the collision.
- Drivers License State: State issuing the driver's license.
- Vehicle ID: Unique identifier for each vehicle involved in the collision.
- Vehicle Damage Extent: Extent of damage to the vehicle.
- Vehicle First Impact Location: Location on the vehicle where the first impact occurred.
- Vehicle Second Impact Location: Location on the vehicle where the second impact occurred.
- Vehicle Body Type: Body type of the vehicle involved.
- Vehicle Movement: Movement of the vehicle at the time of the collision.
- Vehicle Continuing Dir: Direction in which the vehicle continued after the collision.
- Vehicle Going Dir: Direction in which the vehicle was traveling at the time of the collision.
- Speed Limit: Speed limit on the road where the collision occurred.
- Driverless Vehicle: Whether the vehicle involved was driverless.
- Parked Vehicle: Whether the vehicle involved was parked.
- Vehicle Year: Year of manufacture of the vehicle.
- Vehicle Make: Make or manufacturer of the vehicle.
- Vehicle Model: Model of the vehicle.
- Equipment Problems: Any equipment problems reported with the vehicle.
- Latitude: Latitude coordinate of the collision location.
- Longitude: Longitude coordinate of the collision location.
- Location: Combined latitude and longitude coordinates of the collision location.

The data was collected from the Crash Reporting - Drivers Data dataset on Data.gov. It's sourced from the Maryland State Police's Automated Crash Reporting System and reported by local law enforcement in Montgomery County. You can access it on the [Data.gov](https://catalog.data.gov/dataset/crash-reporting-drivers-data) website.

### 2.6 Data Pre-processing and Cleaning

```python
selected_columns = ['Agency Name', 'ACRS Report Type','Crash Date/Time','Route Type','Road Name', 'Cross-Street Type',
       'Cross-Street Name','Related Non-Motorist', 'Collision Type', 'Weather',
       'Surface Condition', 'Light','Traffic Control','Driver Substance Abuse','Driver At Fault', 
       'Injury Severity', 'Circumstance','Driver Distracted By','Vehicle Damage Extent', 'Vehicle First Impact Location','Speed Limit','Vehicle Year',
       'Vehicle Make', 'Vehicle Model', 'Equipment Problems', 'Latitude',
       'Longitude', 'Location']
df1 = df[selected_columns]
```

We have chosen to focus solely on these columns out of all of the columns.

```python
df1.head(3)
```

first three rows after selection

```python
# Replacing missing values in 'Collision Type' with the most common value
df1['Collision Type'].fillna(df['Collision Type'].mode()[0], inplace=True)

# Replacing missing values in 'Weather' and 'Surface Condition' with their respective modes
df1['Weather'].fillna(df['Weather'].mode()[0], inplace=True)

df1['Surface Condition'].fillna(df['Surface Condition'].mode()[0], inplace=True)

# Replacing missing values in 'Traffic Control' with 'Unknown' or 'Not Specified'
df1['Traffic Control'].fillna('Unknown', inplace=True)

# Replacing missing values in 'Municipality' with a placeholder value like 'Unknown'
df1['Vehicle Model'].fillna(df['Vehicle Model'].mode()[0], inplace=True)

#converting date and time to  normal format
df1['Crash Date/Time'] = pd.to_datetime(df1['Crash Date/Time'])

# Converting object-type columns to strings
object_columns = df1.select_dtypes(include=['object']).columns
df1[object_columns] = df[object_columns].astype(str)
```

Because this data is human-written and contains numerous errors, the code below replaces invalid names with names that have similar values.


```python
import pandas as pd

# Dictionary mapping incorrect names to correct ones
corrections = {
    'TOYT': 'TOYOTA',
    'GILL': 'GILLIG',
    'SUBA': 'SUBARU',
    'VOLK': 'VOLKSWAGEN',
    'INFI': 'INFINITI',
    'TOYTA': 'TOYOTA',
    'ACUR': 'ACURA',
    'HYUN': 'HYUNDAI',
    'FREIGHT': 'FREIGHTLINER',
    'HOND': 'HONDA',
    'CHEVROLET': 'CHEVY',
    'CHEVROLET': 'CHEVY',
    'MERZ': 'MERCEDES',
    'CHRYSLERER': 'CHRYSLER',
    'DODG': 'DODGE',
    'PETERBUILT 389': 'PETERBILT 389',
    'GILG': 'GILLIG',
    'HOND': 'HONDA',
    'INFINITI': 'INFINITY',
    'LEX':'LEXUS',
    'CHEV':'CHEVY'
}

# Replace incorrect names with correct ones
df1['Vehicle Make'] = df1['Vehicle Make'].replace(corrections)

# Check the unique values to ensure corrections are applied
print(df1['Vehicle Make'].unique())
```

Using the replace() function and a predefined dictionary (corrections), inconsistent or incorrect values are corrected to ensure uniformity and accuracy in the dataset for vehicle make column.

### 3.3 Waffle chart for distribution of accidents according to "Route type"

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
column_to_visualize = "Route Type"
# Filter out rows with the specified value in the selected column
filtered_df = sample_df[sample_df[column_to_visualize] != "nan"]

# Count the occurrences of each category in the filtered column
category_counts = filtered_df[column_to_visualize].value_counts()

# Calculate the percentage of each category
category_percentages = (category_counts / category_counts.sum()) * 100

# Create the waffle chart
waf = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=category_percentages,
    title={'label': f'Distribution of accidents according{column_to_visualize}', 'loc': 'center', 'size': 22},
    labels=["{0} ({1}%)".format(k, round(v, 2)) for k, v in category_percentages.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.15), 'ncol': len(category_counts), 'framealpha': 0},
    starting_location='NW',
    figsize=(22, 8)
)

# Set background color
waf.gca().set_facecolor('#EEEEEE')
waf.set_facecolor('#EEEEEE')

# Export the visualization
waf.savefig('factor_accidents_waffle.png', dpi=500)

# Show the waffle chart
plt.show()
```


The waffle chart illustrating the distribution of accidents by route type reveals significant insights into the occurrence of accidents across different road categories. Notably, Maryland state roads stand out as the most frequent location for accidents, constituting a substantial proportion of the total incidents at 44.93%. Conversely, service roads experience the lowest incidence of accidents, representing a mere 0.02% of the total. This visualization effectively highlights the disparity in accident rates across various types of routes, emphasizing the prominence of Maryland state roads in accident occurrence while underscoring the relatively low frequency of accidents on service roads.

### 3.4 Word cloud chart for "vehicle make"

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all vehicle makes into a single string
text = ' '.join(sample_df['Vehicle Make'])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Vehicle Makes')
plt.axis('off')
plt.show()
```

In the word cloud chart depicting vehicle makes, Toyota emerges as the most prominent company, indicating its prevalence in the dataset. This observation underscores Toyota's significant presence in reported accidents compared to other car manufacturers. Similarly, Honda, Chevy, and Ford also stand out prominently in the word cloud, suggesting their frequent occurrence in recorded accidents. This insight offers valuable information about the distribution of accidents across different vehicle brands, with Toyota notably leading in frequency within the dataset.
